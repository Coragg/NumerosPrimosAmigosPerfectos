# Números primos

Un número primo, es todo aquel que puede ser dividido por 1 y sigo mismo, si hay otro número que le divida, se le ha de considerar un número compuesto.

```python
# importación de librerias
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def is_prime_number(number: int) -> bool:
    for evaluate_possible_divisible in range(2, number):
        if number % evaluate_possible_divisible == 0:
            return False
    return True
```

```python
def is_prime_number_sqrt(number: int) -> bool:
    sqrt_number = np.sqrt(number)
    counter = 2
    while counter <= sqrt_number:
        if number % counter == 0:
            return False
        counter += 1
    return True
```

## Evaluación de resultados 

En este aparto solo se evaluara si los resultados otorgados por cada algoritmo son los requeridos y si se cumplen, cual es su nivel de eficiencia en la ejecución de sus operaciones. Por lo que se han de seguir los siguietes pasos:

* Analizar mil números para ver si los resultados son los correspondientes
* Tiempo de ejecución en operaciones de 100 mil numeros a un millones
* Graficos en la comparativa de los resultados
* Conclusiones

### Evaluacion de mil numeros

```python
with_sqrt = [is_prime_number_sqrt(n) for n in range(1000)]
without_sqrt = [is_prime_number(n) for n in range(1000)]
```

```python
if with_sqrt == without_sqrt:
    print("Ambos algortimos llegan a los mismos resultados.")
else:
    print("Ambos algoritmos, no llegan a los mismos resultados.")
```

**Estos resultados, nos indican que ambos algoritmos llegan a los mismos resultados, por lo que ahora solo corresponde evaluar su ejecución.**

## Prueba de velocidad con rangos 

```python
def get_numbers(range_of_numbers: int, function_use) -> list:
    return [function_use(n) for n in range(range_of_numbers)]
```

```python
list_of_range = [100, 500, 1000, 10000, 100000, 100000]
```

```python
def execution(range_number: list, function_to_use) -> list:
    time_list: list = []
    for number in range_number:
        start_time = time.time()
        new_list = get_numbers(number, function_to_use)
        end_time = time.time()
        time_list.append(round(end_time - start_time, 3))
    return time_list
```

```python
time_of_prime_number = execution(list_of_range, is_prime_number)
time_of_prime_number_sqrt = execution(list_of_range, is_prime_number_sqrt)
```

```python
time_of_prime_number
```

```python
time_of_prime_number_sqrt
```

## Implementación de lo graficos:

```python
plt.plot(list_of_range, time_of_prime_number, label='Datos sin la aplicacion de la raiz')
plt.plot(list_of_range, time_of_prime_number_sqrt, label='Datos con la aplicacion de la raiz')
plt.title('Comparación de Algoritmos') 
plt.xlabel('Rango de números') 
plt.ylabel('Tiempo en ejecución') 
plt.legend()
```

## Conclusión

Se puede ver que el uso de la raíz cuadra para analizar si un número es primo o compuesto tiene una gran mejorar, que usar el algoritmo en su estado puro.

# Números Amigos
Estos son dos números, donde la suma de un número $a$, da como resultado $b$, y la suma de los divisores de $b$ da como resultado $a$. 

```python
def add_divisors(number: int) -> int:
    sumatory = 0
    for n in range(1, number):
        if number % n == 0:
            sumatory += n
    return sumatory
```

```python
def friendly_numbers(a: int, b: int) -> bool:
    return add_divisors(a) == b and add_divisors(b) == a
```

# Números Perfectos

Un número perfecto, es aquel el cual la suma de sus divisores es el mismo número. Los dos números perfectos más conocidos son $6$ y $28$.

```python
def perfect_number(number: int) -> bool:
    return add_divisors(number) == number
```
